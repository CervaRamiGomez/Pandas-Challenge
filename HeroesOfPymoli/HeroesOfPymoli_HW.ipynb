```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
purchase_data
```

```python
#count total number of players
total_players = purchase_data['SN'].nunique()
total_players_df = pd.DataFrame(
    {"Total Players":[total_players]
    }
)
total_players_df
```

```python
# Number of uniqieItems
unique_items = purchase_data['Item Name'].nunique()

# Avreage Price
average_price = purchase_data['Price'].mean()

# number of Purchases
purchase_count = purchase_data['Price'].count()

# Total Revenue
total_revenue = purchase_data['Price'].sum()

# Print table
purchase_analysis = pd.DataFrame(
    {"Number of Unique Items":[unique_items],
    'Average Price':[average_price],
    'Number of Purchases':[purchase_count],
    'Total Revenue':[total_revenue]
    }
)
purchase_analysis
```

```python
# remove duplictes
individual_count =purchase_data.drop_duplicates('SN')

total_gender_count = individual_count['Gender'].count()
# of Gender count
gender_count = individual_count['Gender'].value_counts()
# gender_count

# percentage of male count
gender_percentage = (gender_count/total_gender_count)*100
#gender_percentage

# set table up
gender_demographics = pd.DataFrame(
    {"Total Count":gender_count,
    'Percentage of Players':gender_percentage
    }
)
# change format to %
gender_demographics["Percentage of Players"] = gender_demographics["Percentage of Players"].astype(float).map("{:,.2f}%".format)
 
#print table
gender_demographics
```

```python

# count purchses per gender
purchase_count = purchase_data.groupby("Gender")["Purchase ID"].count()

# average purchses per gender
average_price = purchase_data.groupby("Gender")["Price"].mean()

# total purchase per gender
total_purchase = purchase_data.groupby("Gender")["Price"].sum()

#print(person)
person = purchase_data.groupby("Gender")["SN"].nunique()

#print(price_per_person)
price_per_person = total_purchase/person

# set table
gender_p_a = pd.DataFrame(
    {"Purchase Count": purchase_count,
    "Average Purchase Price": average_price,
    "Total Purchase Value": total_purchase,
    "Avg Total Purchase per Person": price_per_person}
)
# format columns
#gender_p_a["Average Purchase Price"] = gender_p_a["Average Purchase Price"].astype(float).map("${:,.2f}".format)
#gender_p_a["Total Purchase Value"] = gender_p_a["Total Purchase Value"].astype(float).map("${:,.2f}".format)
#gender_p_a["Avg Total Purchase per Person"] = gender_p_a["Avg Total Purchase per Person"].astype(float).map("${:,.2f}".format)

# print table
gender_p_a
```

```python
# Figure out the minimum and maximum
# print(purchase_data['Age'].max())
# print(purchase_data['Age'].min())

# Create bins in which to place values 
bins = [0,9,14,19,24,29,34,39,45]

# Create labels for these bins
group_labels = ['<10','10-14','15-19','20-24','25-29','30-34','35-39','40+']

# Slice the data and place it into bins
pd.cut(purchase_data['Age'], bins, labels=group_labels).head()

# Place the data series into a new column inside of the DataFrame
purchase_data['Total Count'] = pd.cut(purchase_data["Age"], bins, labels=group_labels)

purchase_data_age = purchase_data.groupby('Total Count')
#print(purchase_data_age['Total Count'].count)
#purchase_data_age[["Total Count", "Percentage of Players"]].mean()



```

```python
# Most Popular Items

# obtain purchase count
purchase_count = purchase_data.groupby("SN")["Purchase ID"].count()

# Obtain average purchse Price
purchase_price = purchase_data.groupby("SN")["Price"].mean()

# Obtain total purchase value
total_purchase_value = purchase_data.groupby("SN")["Price"].sum()

#set up table
players_df = pd.DataFrame(
    {
    "Purchase Count": purchase_count,
    "Average Purchase Price": purchase_price,
    "Total Purchase Value": total_purchase_value
    
    }
)

# format table Values
players_df["Average Purchase Price"] = players_df["Average Purchase Price"].astype(float).map("${:,.2f}".format)
players_df["Total Purchase Value"] = players_df["Total Purchase Value"].astype(float).map("${:,.2f}".format)

# print table
players_df

players_df = players_df.sort_values(["Purchase Count"], ascending=False)
players_df.head(5)
```

```python
# Counting Purchases
purchase_count = purchase_data.groupby(["Item ID","Item Name"])["Purchase ID"].count()

# Identifing Item Price
item_price = purchase_data.groupby(["Item ID","Item Name"])["Price"].mean()

# Purchace Value
total_purchase_value = purchase_data.groupby(["Item ID","Item Name"])["Price"].sum()

# Set table
most_popular_df = pd.DataFrame({"Purchase Count": purchase_count,
                                "Item Price": item_price,
                                "Total Purchase Value": total_purchase_value})   
# change Format                                                            
most_popular_df["Item Price"] = most_popular_df["Item Price"].astype(float).map("${:,.2f}".format)
most_popular_df["Total Purchase Value"] = most_popular_df["Total Purchase Value"].astype(float).map("${:,.2f}".format)

#organize Data
most_popular_df = most_popular_df.sort_values("Purchase Count", ascending= False)

# print table
most_popular_df.head()
```
