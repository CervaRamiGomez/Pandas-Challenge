```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
#school_data_complete
```

```python
# District Summary #

# Calculate Total schools
total_schools_df = school_data_complete["school_name"].nunique()
# print(total_schools_df)

# Calculate Total students
total_students_df = school_data_complete["student_name"].count()
# print(total_students_df)

# Calculate Total budget
total_budget_df = school_data["budget"].sum()
# print(total_budget_df)

# Calculate Average math score
average_math_score = school_data_complete["math_score"].mean()
# print(average_math_score)

# Calculate Average reading score
average_reading_score = school_data_complete["reading_score"].mean()
#print(average_reading_score)

# Calculate % passing math (the percentage of students who passed math)
passing_math_percentage = (school_data_complete[school_data_complete["math_score"] >= 70].math_score.count())/total_students_df*100
# print(passing_math_percentage)

# Calculate % passing reading (the percentage of students who passed reading)
passing_reading_percentage = (school_data_complete[school_data_complete["reading_score"] >= 70].reading_score.count())/total_students_df*100
# print(passing_reading_percentage)

# Calculate % overall passing (the percentage of students who passed math AND reading)
passing_reading_and_math_percentage = ((school_data_complete[(school_data_complete["math_score"] >= 70) & (school_data_complete["reading_score"] >= 70)].math_score.count())/total_students_df)*100

# lay out table (passing_reading_and_math_percentage)
district_summary_df = pd.DataFrame(
    {"Total Schools": [total_schools_df],
    "Total Students": [total_students_df],
    "Total Budget": [total_budget_df],
    "Average Math Score": [average_math_score],
    "Average Reading Score": [average_reading_score],
    "% Passing Math": [passing_math_percentage],
    "% Passing Reading": [passing_reading_percentage],
    "% Passing Overall":[passing_reading_and_math_percentage]}
)
# change Total Budget to currency
district_summary_df["Total Budget"] = district_summary_df["Total Budget"].astype(float).map("${:,.2f}".format)
# Print table
district_summary_df
```

```python
# School Summary # groupby

# Calculate School name
school_names_summary = school_data_complete.groupby('school_name')
# print(school_names_summary)

# Calculate School type
school_type_summary = school_data_complete.groupby('school_name')['type'].first()
# print(school_type_summary)

# Calculate Total students
total_student_count_per_school = school_data_complete.groupby('school_name')['size'].first()
# total_student_count_per_school.head()

# Calculate Total school budget
total_school_budget = school_data_complete.groupby('school_name')['budget'].first()
# total_school_budget.head()

# Calculate Per student budget
total_student_budget = total_school_budget/total_student_count_per_school
#total_student_budget.head()

# Calculate Average math score
schools_average_math_score = school_data_complete.groupby('school_name')['math_score'].mean()
#schools_average_math_score.head()

# Calculate Average reading score
schools_average_reading_score = school_data_complete.groupby('school_name')['reading_score'].mean()
#schools_average_reading_score.head()

# Calculate % passing math (the percentage of students who passed math)
schools_passing_math_percentage = ((school_data_complete[school_data_complete["math_score"] >= 70].groupby('school_name').math_score.count())/total_student_count_per_school)*100
#schools_passing_math_percentage.head()

# Calculate % passing reading (the percentage of students who passed reading)
schools_passing_reading_percentage = ((school_data_complete[school_data_complete["reading_score"] >= 70].groupby('school_name').reading_score.count())/total_student_count_per_school)*100
#schools_passing_reading_percentage.head()

# Calculate % overall passing (the percentage of students who passed math AND reading)
schools_passing_math_and_reading_percentage =((school_data_complete[(school_data_complete["math_score"] >= 70) & (school_data_complete["reading_score"] >= 70)].groupby("school_name").math_score.count())/total_student_count_per_school)*100
#schools_passing_math_and_reading_percentage.head()

# lay out table (passing_reading_and_math_percentage)
schools_summary_df = pd.DataFrame(
    {"School Type": school_type_summary,
    "Total Students": total_student_count_per_school,
    "Total School Budget": total_school_budget,
    "Per Student Budget": total_student_budget,
    "Average Math Score": schools_average_reading_score,
    "Average Reading Score": schools_average_reading_score,
    "% Passing Math": schools_passing_math_percentage,
    "% Passing Reading":schools_passing_reading_percentage,
    "% Overall Passing":schools_passing_math_and_reading_percentage}
)
# change Total Budget to currency
schools_summary_df["Total School Budget"] = schools_summary_df["Total School Budget"].astype(float).map("${:,.2f}".format)
schools_summary_df["Per Student Budget"] = schools_summary_df["Per Student Budget"].astype(float).map("${:,.2f}".format)

# Print table
schools_summary_df

```

```python
# Highest-Performing Schools (by % Overall Passing) #
top_five_school_summary_df = schools_summary_df.sort_values(["% Overall Passing"], ascending=False)
top_five_school_summary_df.head(5)
```
